```python
import cv2
from IPython.display import Image
```

```python
img_url = "Imagens/faces3.jpg"

Image(url=img_url)
```

```python
img_face = cv2.imread(img_url)
img_face.shape
```

```python
img_face = cv2.imread(img_url)
img_face_gray = cv2.cvtColor(img_face, cv2.COLOR_BGR2GRAY)
```

```python
print(img_face_gray.shape)
```

```python
face_parametro = "Cascade/frontalface_alt.xml"
classificador_cascade = cv2.CascadeClassifier(face_parametro)
```

```python
classificador_cascade
```

```python
faces = classificador_cascade.detectMultiScale(
    img_face_gray, 
    scaleFactor=1.2, 
    minNeighbors=10, 
    minSize=(30, 30)
)
```

```python
faces
```

```python
for (x, y, width, height) in faces:
    cv2.rectangle(img_face, (x, y), (x +width, y+height), (0, 255, 0), 6 )
```

```python
cv2.imwrite("Imagens/faces_resultado.jpg", img_face)
```
